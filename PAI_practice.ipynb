#1. Water Jug Problem

```python
from collections import deque

def WaterJugBFS(jug1,jug2,target):
  #if not possible return karo
  if target > max(jug1,jug2):
    return False
  #create a set to track visited nodes
  visited = set()
  visited.add((0,0))
  #create a parent mapping to map parent node of every visited node
  parent = {(0,0):None}
  #finally a queue to run BFS
  queue = deque()
  queue.append((0,0))
  #let's loop
  while queue:
    #Extract the left most element
    current = queue.popleft()
    a,b=current
    #inspect if it is solution
    if a==target or b==target:
      #if yes, extract path and return
      path = []
      while current is not None:
        path.append(current)
        current = parent[current]
      path.reverse()
      return path
    #else we geenrate all children of the current state and go on searching!
    generated = [(jug1,b),(a,jug2),(0,b),(a,0),(max(0, a-(jug2-b)), min(b+a, jug2)), (min(a+b,jug1), max(0, b-(jug1-a)))]
    for next_state in generated:
        if next_state not in visited:
          visited.add(next_state)
          queue.append(next_state)
          parent[next_state]=current
  return False


WaterJugBFS(5,7,4)
```

```python
def WaterJugDFS(jug1,jug2,target):
  visited = set()
  path = []

  def dfs(a,b):
    if (a,b) in visited:
      return False
    visited.add((a,b))
    path.append((a,b))
    if a == target:
      path.append((a,0))
      return True
    if b==target:
      path.append((0,b))
      return True
    if dfs(0,b):
      return True
    if dfs(jug1,b):
      return True
    if dfs(a,0):
      return True
    if dfs(a,jug2):
      return True
    if dfs(max(0, a-(jug2-b)), min(b+a, jug2)):
      return True
    if dfs(min(a+b,jug1), max(0, b-(jug1-a))):
      return True
    path.pop()
    return False

  if dfs(0,0):
    return path
  return False

WaterJugDFS(5,7,4)
```

```python
#Water jug using A star
import heapq
def WaterJugAstar(jug1, jug2, goal):
  def is_goal(state):
    return state[0]==goal
  def heuristic(state):
    return abs(state[0]-goal)
  def get_neighbours(state):
    a,b = state
    neighbours = [(jug1,b),(a,jug2),(0,b),(a,0),(max(0, a-(jug2-b)), min(b+a, jug2)), (min(a+b,jug1), max(0, b-(jug1-a)))]
    return neighbours

  visited = set()
  pq = []
  start = (0,0)
  heapq.heappush(pq, (heuristic(start),0,start, [start]))

  while pq:
    f, g, current, path = heapq.heappop(pq)
    if current in visited:
      continue
    visited.add(current)

    if is_goal(current):
      return path

    for neighbour in get_neighbours(current):
      if neighbour not in visited:
        g = g+1
        f = g+heuristic(neighbour)
        heapq.heappush(pq, (f,g,neighbour, path+[neighbour]))
  return None

print(WaterJugAstar(4,3,2))
```

#A-star algorithm

```python
# Graph represented as an adjacency list with edge costs
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 5, 'E': 12},
    'C': {'F': 2},
    'D': {'G': 3},
    'E': {'G': 1},
    'F': {'E': 3},
    'G': {}
}

# Heuristic values for each node (straight-line distance to goal 'G')
h = {
    'A': 10,
    'B': 8,
    'C': 5,
    'D': 7,
    'E': 3,
    'F': 6,
    'G': 0
}


def Astar_search(start, stop):
  open_set = [start]
  closed_set= set()
  parents={}
  g={}
  g[start]=0
  parents[start]=None

  while open_set:
    current = None
    for node in open_set:
      if current==None or g[node]+h[node]<g[current]+h[current]:
        current = node
    if current is None:
      return None
    if current == stop:
      path=[]
      while parents[current]!=None:
        path.append(current)
        current = parents[current]
      path.append(start)
      path.reverse()
      return path
    open_set.remove(current)
    closed_set.add(current)
    if graph.get(current) is None:
      continue
    for (n,w) in graph.get(current):
      if n in closed_set:
        continue
      if n not in open_set:
        parents[n]=current
        g[n]=g[current]+w
        open_set.append(n)
      else:
        if g[n]>g[n]+w:
          g[n]=g[n]+w
          parents[n]=current
          if n in closed_set:
            closed_set.remove(n)
            open_set.append(n)
  return False
```

#8-puzzle solver

```python
import heapq
goal=[1,2,3,4,5,6,7,8,0]

def heuristic(state):
  distance=0
  for i in range(9):
    if state[i]!=0:
      goal_pos=goal.index(state[i])
      distance+=abs(i//3-goal_pos//3)+abs(i%3-goal_pos%3)
  return distance

def next_states(state):
  moves = []
  idx=state.index(0)
  row, col = idx//3, idx%3
  directions = [(-1,0),(1,0),(0,1),(0,-1)]
  for dr, dc in directions:
    new_row, new_col = row+dr, col+dc
    if 0<=new_row<3 and 0<new_col<3:
      new_idx = new_row*3+new_col
      new_state = state[:]
      new_state[new_idx], new_state[idx] =new_state[idx], new_state[new_idx]
      moves.append(new_state)
  return moves


def a_star(initial):
  open_list = []
  heapq.heappush(open_list, (heuristic(initial),0,initial,[]))

  visited = set()
  visited.add(tuple(initial))

  while open_list:
    f,g,current_state, path = heapq.heappop(open_list)
    if current_state == goal:
      return path + [current_state]
    for neighbour in next_states(current_state):
      if tuple(neighbour) not in visited:
        visited.add(tuple(neighbour))
        h = heuristic(neighbour)
        heapq.heappush(open_list, (g+h+1,g+1, neighbour, path+[current_state]))
  return None

def format_state(state):
  return [state[i:i+3] for i in range(0,9,3)]

start = [1,2,3,4,8,0,7,6,5]

solution = a_star(start)

if solution:
  print("solution:")
  for i in solution:
    for row in format_state(i):
      print(row)
    print()
else:
  print("NO SOLUTION FOUND!")
```

#CSP

```python
def solve_n_queens(n):
  board = [['.' for _ in range(n)] for _ in range(n)]
  solutions = []

  def is_safe(row, col):
    #check for same column
    for i in range(row):
      if board[i][col] == 'Q':
        return False

    #check  diagonals
    i,j = row-1, col-1
    while i>=0 and j>=0:
      if board[i][j] == 'Q':
        return False
      i-=1
      j-=1

    i,j = row-1, col+1
    while i>=0 and j<n:
      if board[i][j]=='Q':
        return False
      i-=1
      j+=1
    return True

  def backtrack(row):
    if row==n:
      solutions.append([''.join(r) for r in board])
      return
    for col in range(n):
      if is_safe(row, col):
        board[row][col] = 'Q'
        backtrack(row+1)
        board[row][col] = '.'
  backtrack(0)
  return solutions
n=int(input("Enter n for N-Queens"))
solutions = solve_n_queens(n)
for i in range(len(solutions)):
    print(f"solution {i+1}")
    for row in solutions[i]:
      print(row)
    print()
```

#GAME SEARCH

```python
#mini max search
import math
def minimax(current_depth, node_index, max_turn, scores, target_depth):
  if current_depth == target_depth:
    return scores[node_index]
  if max_turn:
    return max(minimax(current_depth+1, node_index*2, False, scores, target_depth), minimax(current_depth+1, node_index*2+1, False, scores, target_depth))
  else:
    return min(minimax(current_depth+1, node_index*2, True, scores, target_depth), minimax(current_depth+1, node_index*2+1, True, scores, target_depth))

scores=[3,5,2,9,12,5,23,23]
tree_depth = int(math.log2(len(scores)))
print("The optimal value is:", minimax(0,0,True,scores,tree_depth))
```

```python
MAX, MIN = float('inf'), float('-inf')
def aplhabetapruning(depth, node_index, max_turn, scores,target_depth, alpha, beta):
  if depth == target_depth:
    return scores[node_index]
  if max_turn:
    best = MIN
    for i in range(2):
      val = aplhabetapruning(depth+1, node_index*2+i, False, scores, target_depth, alpha, beta)
      best = max(best, val)
      alpha = max(best, alpha)
      if beta <= alpha:
        break
    return best
  else:
    best = MAX
    for i in range(2):
      val = aplhabetapruning(depth+1, node_index*2+i, True, scores, target_depth, alpha, beta)
      best = min(best, val)
      beta = min(beta, best)
      if beta<=alpha:
        break
    return best

scores = [3,5,6,9,1,2,0,7]
target_depth=math.log2(len(scores))
print("Optimal value is:", aplhabetapruning(0,0,True,scores,target_depth, MIN, MAX))
```

```python
import numpy as np

grid = [[0,0,0,1],
        [0,None,0,-1],
        [0,0,0,0]]

rows, cols = len(grid), len(grid[0])
default_reward = -0.1
gamma = 0.9
theta = 1e-3

actions=['U','L','R','D']

action_delta = {
    'U':(-1,0),
    'L':(0,-1),
    'R':(0,1),
    'D':(1,0)
}

transitions = {
    'U':[('U',0.8), ('L',0.1), ('R',0.1)],
    'L':[('U',0.1), ('L',0.8), ('D',0.1)],
    'R':[('U',0.1), ('R',0.8), ('D',0.1)],
    'D':[('L',0.1), ('R',0.1), ('D',0.8)]
}

def is_valid(r,c):
  return 0<=r<rows and 0<=c<cols and grid[r][c] is not None

def get_reward(r,c):
  return grid[r][c] if grid[r][c] is not None else default_reward
def is_terminal(r,c):
  return grid[r][c] in [1,-1]
```

```python
def value_iteration():
    V = np.zeros((rows, cols))
    policy = [['' for _ in range(cols)] for _ in range(rows)]
    iteration = 0

    while True:
        delta = 0
        new_V = np.copy(V)

        print(f"\nIteration {iteration}:")
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] is None:
                    continue  # Skip wall

                if is_terminal(r, c):
                    new_V[r][c] = grid[r][c]
                    continue  # Terminal state

                best_value = float('-inf')
                best_action = None

                for a in actions:
                    total = 0
                    for move, prob in transitions[a]:
                        dr, dc = action_delta[move]
                        nr, nc = r + dr, c + dc

                        if not is_valid(nr, nc):
                            nr, nc = r, c  # stay in place if invalid move

                        reward = grid[nr][nc] if is_terminal(nr, nc) else default_reward
                        total += prob * (reward + gamma * V[nr][nc])

                    if total > best_value:
                        best_value = total
                        best_action = a

                new_V[r][c] = best_value
                policy[r][c] = best_action
                delta = max(delta, abs(V[r][c] - best_value))

        V = new_V
        print(np.round(V, 2))  # Print current value function

        iteration += 1
        if delta < theta:
            break

    return V, policy

# Run value iteration
V, policy = value_iteration()

# Print final results
print("\nFinal Value Function:")
print(np.round(V, 2))

print("\nFinal Policy:")
for row in policy:
    print(' '.join([a if a else '.' for a in row]))
```

```python
#Policy Iternation
import numpy as np

# Grid configuration
grid = [
    [0,  0,   0,  +1],
    [0, None, 0,  -1],
    [0,  0,   0,   0]
]

rows, cols = len(grid), len(grid[0])

# Rewards
default_reward = -0.1
gamma = 0.9
theta = 1e-4

actions = ['U', 'L', 'D', 'R']
action_delta = {
    'U': (-1, 0),
    'D': (1, 0),
    'L': (0, -1),
    'R': (0, 1)
}

transitions = {
    'U': [('U', 0.8), ('L', 0.1), ('R', 0.1)],
    'D': [('D', 0.8), ('R', 0.1), ('L', 0.1)],
    'L': [('L', 0.8), ('D', 0.1), ('U', 0.1)],
    'R': [('R', 0.8), ('U', 0.1), ('D', 0.1)]
}

def is_valid(r, c):
    return 0 <= r < rows and 0 <= c < cols and grid[r][c] is not None

def is_terminal(r, c):
    return grid[r][c] in [+1, -1]

def policy_evaluation(policy, V):
    while True:
        delta = 0
        new_V = np.copy(V)

        for r in range(rows):
            for c in range(cols):
                if grid[r][c] is None or is_terminal(r, c):
                    continue

                a = policy[r][c]
                total = 0
                for move, prob in transitions[a]:
                    dr, dc = action_delta[move]
                    nr, nc = r + dr, c + dc

                    if not is_valid(nr, nc):
                        nr, nc = r, c

                    reward = grid[nr][nc] if is_terminal(nr, nc) else default_reward
                    total += prob * (reward + gamma * V[nr][nc])

                new_V[r][c] = total
                delta = max(delta, abs(V[r][c] - total))

        V = new_V
        if delta < theta:
            break
    return V

def policy_iteration():
    V = np.zeros((rows, cols))
    policy = [['U' if grid[r][c] is not None and not is_terminal(r, c) else '' for c in range(cols)] for r in range(rows)]

    iteration = 0
    while True:
        print(f"\nPolicy Iteration {iteration}:")
        V = policy_evaluation(policy, V)

        policy_stable = True
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] is None or is_terminal(r, c):
                    continue

                old_action = policy[r][c]
                best_action = old_action
                best_value = float('-inf')

                for a in actions:
                    total = 0
                    for move, prob in transitions[a]:
                        dr, dc = action_delta[move]
                        nr, nc = r + dr, c + dc

                        if not is_valid(nr, nc):
                            nr, nc = r, c

                        reward = grid[nr][nc] if is_terminal(nr, nc) else default_reward
                        total += prob * (reward + gamma * V[nr][nc])

                    if total > best_value:
                        best_value = total
                        best_action = a

                if best_action != old_action:
                    policy_stable = False
                    policy[r][c] = best_action

        print(np.round(V, 2))

        iteration += 1
        if policy_stable:
            break

    return V, policy

# Run policy iteration
V, policy = policy_iteration()

# Print final results
print("\nFinal Value Function:")
print(np.round(V, 2))

print("\nFinal Policy:")
for row in policy:
    print(' '.join([a if a else '.' for a in row]))
```
